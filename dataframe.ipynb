# Penalty Success Rate

```python
df.columns
```

```python
penalty=df[['Player','PK','PKatt']]
penalty
```

```python
penalty=penalty.sort_values('PK',ascending=False)
pd1=penalty.head(14)
```

```python
fig=plt.figure(figsize=(10,8))
Players=list(pd1['Player'])[::-1]
penalty=list(pd1['PK'])[::-1]
penaltyattms=list(pd1['PKatt'])[::-1]
ax1=fig.add_subplot()

ax1.barh(Players,penaltyattms,hatch='.',height=0.85)
ax1.barh(Players,penalty,hatch='/',height=0.85)

plt.show()
```
