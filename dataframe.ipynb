# Penalty Success Rate

```python
df.columns
```

```python
penalty=df[['Player','PK','PKatt']]
penalty
```

```python
penalty=penalty.sort_values('PK',ascending=False)
pd1=penalty.head(14)
```

**Player with Maximum penalty goals scored**

```python
fig=plt.figure(figsize=(10,8))
Players=list(pd1['Player'])[::-1]
penalty=list(pd1['PK'])[::-1]
penaltyattms=list(pd1['PKatt'])[::-1]
ax1=fig.add_subplot()

ax1.set_xticks(np.arange(0,max(penaltyattms)+1))

ax1.barh(Players,penaltyattms,hatch='.',height=0.85,label="Penalty Attempted")
ax1.barh(Players,penalty,hatch='.',height=0.85,label="Penalty Scored")

ax1.set_xlabel("Pentaly Attempted and Scored",fontsize=16,fontdict={'fontname':'Century Gothic','weight':'bold'})
ax1.set_ylabel("Players",fontsize=16,fontdict={'fontname':'Century Gothic','weight':'bold'})
ax1.set_title("Penalty Attempted and Scored by Players from top-5 Leagues (3/26/2025)",fontsize=18,fontdict={'fontname':'Century Gothic','weight':'bold'},loc='right')

plt.legend(prop={'size':14,'family':'Century Gothic','weight':'bold'})
# plt.grid(axis='x',alpha=0.8,linestyle='--',color='#000000')
plt.show()
```
